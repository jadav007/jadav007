```python
import pandas as pd
import numpy as np

# Sample demographic data
df =pd.DataFrame({
    'employee_id':['B182400','B182410','B182420','B182430','B182440','B182450','B182460','B182470','B182480','B182490'],
    'Age': [23, 45, 34, 29, 56, 40, 30, 24, 45, 50],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female'],
    'highest_qualification':['Bachelors','Masters','12th','Masters','Bachelors','Masters','Bachelors','Bachelors','Bachelors','11th'],
    'Income': [50000, 60000, 55000, 70000, 65000, 80000, 58000, 49000, 71000, 62000]
})
print(df)
def demographics(df):
    # Mean, median, and standard deviation for Age and Income
    age_mean = df['Age'].mean()
    age_median = df['Age'].median()
    age_std = df['Age'].std()

    income_mean = df['Income'].mean()
    income_median = df['Income'].median()
    income_std = df['Income'].std()

    # Gender distribution
    gender_counts = df['Gender'].value_counts()

    return {
        'Age': {
            'Mean': age_mean,
            'Median': age_median,
            'Standard Deviation': age_std
        },
        'Income': {
            'Mean': income_mean,
            'Median': income_median,
            'Standard Deviation': income_std
        },
        'Gender Distribution': gender_counts.to_dict()
    }

# Analyze the data
statistics = demographics(df)

# Print results
print("Demographic Data Analysis:")
print(f"Age - Mean: {statistics['Age']['Mean']:.2f}")
print(f"Age - Median: {statistics['Age']['Median']:.2f}")
print(f"Age - Standard Deviation: {statistics['Age']['Standard Deviation']:.2f}")
print()
print(f"Income - Mean: {statistics['Income']['Mean']:.2f}")
print(f"Income - Median: {statistics['Income']['Median']:.2f}")
print(f"Income - Standard Deviation: {statistics['Income']['Standard Deviation']:.2f}")
print(f"Persons who are earning more than 50k are:\n{df[df['Income']>50000]}")
print(f"Persons who have qualification as:\n{df['highest_qualification']}")
print(f"Persons ages more than average age:\n{df[df['Age']>df['Age'].mean()]}")

print()
print("Gender Distribution:")
for gender, count in statistics['Gender Distribution'].items():
    print(f"{gender}: {count}")
```
