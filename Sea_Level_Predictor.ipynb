```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Step 1: Import the data
df = pd.read_csv(r'C:\jupyter folder\epa-sea-level.csv')  # Use raw string for file path

# Step 2: Create the scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], color='blue', label='Data Points')

# Step 3: Create a line of best fit for all data
slope_all, intercept_all, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
years_all = pd.Series(range(1880, 2051))  # From 1880 to 2050
plt.plot(years_all, intercept_all + slope_all * years_all, 'r', label='Best Fit Line (All Data)')

# Step 4: Create a line of best fit for data from the year 2000 onward
df_2000 = df[df['Year'] >= 2000]
slope_2000, intercept_2000, r_value_2000, p_value_2000, std_err_2000 = linregress(df_2000['Year'], df_2000['CSIRO Adjusted Sea Level'])
years_2000 = pd.Series(range(2000, 2051))  # From 2000 to 2050
plt.plot(years_2000, intercept_2000 + slope_2000 * years_2000, 'g', label='Best Fit Line (2000 Onwards)')

# Step 5: Add labels and title
plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.legend()
plt.grid()

# Step 6: Save and show the plot
plt.savefig('sea_level_rise.png')
plt.show()
```
